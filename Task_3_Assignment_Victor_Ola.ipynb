## Question 6: Peak Hours of Sales

```python
#To compute the data for peak hours and sales quantity
pizza_order_data["time"] = pd.to_datetime(pizza_order_data["time"], format='%H:%M:%S')
peak_hours = pizza_order_data.groupby(pizza_order_data["time"].dt.hour)["quantity"].sum()
peak_sales_hours = peak_hours.sort_values(ascending = False).index[0]
sales_quantity = peak_hours.sort_values(ascending = False).values[0]

#Plot code to show sales of the day and the peak sales hours
plt.plot(peak_hours, color = "red")
plt.title("Sales in a day")
plt.xlabel("Time of the Day")
plt.ylabel("Quantity Sold")

#inserting grid lines in the graph for easy reading of values
plt.grid(True)

plt.show()

print(f"The peak sales hour as shown in the graph is {peak_sales_hours} noon with cummulative sales quantity of: '{sales_quantity}'")
```

## Question 7: Total Sales on Day of Week and Day with Most Sales

```python
#To compute the data for sales in the week by day and sales quantity
pizza_order_data["date"] = pd.to_datetime(pizza_order_data["date"])
total_sales_day_of_week = pizza_order_data.groupby(pizza_order_data["date"].dt.day_name())["quantity"].sum()

#day_with_most_sales = total_sales_day_of_week.idxmax()
day_with_most_sales = total_sales_day_of_week.sort_values(ascending = False).index[0]
quantity_week = int(total_sales_day_of_week.sort_values(ascending = False).values[0])

#Plot code to show sales in the week by day
plt.bar(total_sales_day_of_week.index, total_sales_day_of_week.values, color = "green")
plt.title("Total Sales in a Week")
plt.xlabel("Day of the Week")
plt.ylabel("Quantity Sold")
plt.xticks(rotation = 90)

plt.show()

#To print the days of the week and their individual sales
total_sales_day_of_week_string = total_sales_day_of_week.to_string().replace("Name: quantity, dtype: float64", "").strip()
print(f"The total sales on each day of the week is:\n{total_sales_day_of_week_string}")
print(f"\nThe day with the most sales is: '{day_with_most_sales}' at a quantity of: {quantity_week}")
```

## Question 8: Top 5 Bestselling Pizzas

```python
#To compute the data top five bestseeling pizzas and sales quantity
top_five_bestselling_pizzas = pizza_order_data.groupby("pizza_type_id")["quantity"].sum().sort_values(ascending = False).head(5)
top_fifteen_bestselling_pizzas = pizza_order_data.groupby("pizza_type_id")["quantity"].sum().sort_values(ascending = False).head(15)

#Plot code to show the five bestseeling pizzas and sales quantity
plt.bar(top_fifteen_bestselling_pizzas.index, top_fifteen_bestselling_pizzas.values, color = "blue")
plt.title("Top 5 Bestselling Pizzas", fontweight = "bold")
plt.xlabel("Pizza Types")
plt.ylabel("Quantity Sold")
plt.xticks(rotation = 90)

plt.show()

# Convert the Series to a string and remove the unwanted part for printing
top_five_bestselling_pizzas_string = top_five_bestselling_pizzas.to_string().replace("Name: quantity, dtype: float64", "").strip()

#To print the five bestselling pizzas
print(f"The top five best selling pizzas are: \n{top_five_bestselling_pizzas_string}")
```

## Question 9: Sales Made in Each Month.

```python
#To compute the data for monthly sales quantity
pizza_order_data["date"] = pd.to_datetime(pizza_order_data["date"])
sales_in_each_month = pizza_order_data.groupby(pizza_order_data["date"].dt.month_name())["quantity"].sum()

#arrange month name for sales in each month in order from january to december
sales_in_each_month = sales_in_each_month.reindex(["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"])

#Plot code to show the monthly sales quantity
plt.bar(sales_in_each_month.index, sales_in_each_month.values, color = "orange")
plt.title("Monthly Sales", fontweight = "bold")
plt.xlabel("Month")
plt.ylabel("Quantity Sold")
plt.xticks(rotation = 90)

plt.show()

# Convert the Series to a string and remove the unwanted part for printing
sales_in_each_month_string = sales_in_each_month.to_string().replace("Name: quantity, dtype: float64", "").strip()

#To print the monthly sales quantity
print(f"The sales made in each month are: \n{sales_in_each_month_string}")
```

## Question 10: Pizza Types Not Doing Well on the menu

```python
pizza_types_not_doing_well = pizza_order_data.groupby("pizza_type_id")["quantity"].sum().sort_values(ascending = True).head(10)

#Plot code to show the monthly sales quantity
plt.bar(pizza_types_not_doing_well.index, pizza_types_not_doing_well.values, color = "brown")
plt.title("Monthly Sales", fontweight = "bold")
plt.xlabel("Month")
plt.ylabel("Quantity Sold")
plt.xticks(rotation = 90)

plt.show()

# Convert the Series to a string and remove the unwanted part for printing
pizza_types_not_doing_well = pizza_types_not_doing_well.to_string().replace("Name: quantity, dtype: float64", "").strip()

#To print the Pizzas no doing well
print(f"The pizza types not doing well are: \n{pizza_types_not_doing_well}")
```
